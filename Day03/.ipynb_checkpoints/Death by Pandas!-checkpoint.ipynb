## Loading a DataFrame

```python
path = '/Users/matthew.mitchell/Documents/''
```

```python
df = pd.read_csv(path+'causes_of_death.tsv', delimiter = '\t')
print(len(df))
df.head(10)
```

```python
df.Population.value_counts()[:10]
```

```python
def remove_not_applicable(x):
    if x == 'Not Applicable':
        return 0
    else:
        return x
```

```python
df['Population'] = df['Population'].map(remove_not_applicable)
df.Population.value_counts()[:10]
```

```python
df[df['Population']!='Not Applicable']
```

```python
grouped = df.groupby('State')['Deaths'].sum().reset_index()
grouped.head(2)
```

```python
grouped = grouped.sort_values(by='Deaths', ascending=False)
grouped.head()
```

```python
grouped = df.groupby(['Gender', 'Race'])['Deaths'].sum().reset_index()
grouped.head()
```

## Pivots
